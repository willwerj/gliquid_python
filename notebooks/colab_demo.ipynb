## 2) Configure API key and data location

```python
import os
from pathlib import Path

# Use your Materials Project API key
os.environ["NEW_MP_API_KEY"] = "YOUR_API_KEY_HERE"

import gliquid.config as cfg

# Keep data outside the package; point GLiquid to the repo data directory
cfg.set_data_dir(Path("/content/gliquid_python/data").resolve())

print(f"Project root : {cfg.project_root}")
print(f"Data directory: {cfg.data_dir}")
```

## 3) Run a quick binary fitting workflow

```python
from gliquid.binary import BinaryLiquid, BLPlotter

bl = BinaryLiquid.from_cache("Cu-Mg", param_format="comb-exp")
fit_results = bl.fit_parameters(verbose=True, n_opts=3)
best_fit = min(fit_results, key=lambda x: x.get("mae", float("inf")), default={})

print("Best-fit result:")
for field, value in best_fit.items():
    print(f"  {field}: {value}")

plotter = BLPlotter(bl)
plotter.show("fit+liq")
```
