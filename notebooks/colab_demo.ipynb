## 3) Run a quick binary fitting workflow\n

```python
from gliquid.binary import BinaryLiquid, BLPlotter\n\nbl = BinaryLiquid.from_cache("Cu-Mg", param_format="comb-exp")\nfit_results = bl.fit_parameters(verbose=True, n_opts=3)\nbest_fit = min(fit_results, key=lambda x: x.get("mae", float("inf")), default={})\n\nprint("Best-fit result:")\nfor field, value in best_fit.items():\n    print(f"  {field}: {value}")\n\nplotter = BLPlotter(bl)\nplotter.show("fit+liq")\n
```
